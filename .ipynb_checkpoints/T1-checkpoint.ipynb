```r
require(dyRank)
require(tidyverse)
require(knitr)
data("f1_race", package = "dyRank")
## load additional data
## see ?f1_grid and ?f1_laptime for the details
```

```r
set.seed(2018)
## estimate via MCMC
fit <- dyRank(
    data       = f1_race,
    var_rank   = "Pos",           
    var_player = "driver",
    var_match  = "GP",
    var_time   = "year",
    driver_fix = "Fernando Alonso",
    mcmc = 100, burnin = 10, thin = 1,
    truncation = 15
)
length(fit$lambda)
length(fit$lambda[[1]])
length(fit$lambda[[1]][[1]])
```

```r
dim(fit$data$dat_ref)
slice_head(as_tibble(fit$data$dat_ref), n = 2000) %>%
    knitr::kable(digits = 3)
```

```r
fit$data$n_drivers
fit$data$n_race
rating <- get_rating(fit)
```

```r
slice_head(as_tibble(rating),n = 10) %>%
    knitr::kable(digits = 5)
```

```r
drivers_onechamp <- c("Kimi Räikkönen","Fernando Alonso","Jenson Button","Nico Rosberg","Lewis Hamilton","Sebastian Vettel","Michael Schumacher","Ayrton Senna","Alain Prost")
onechamp <- plot_rating(rating,facet = TRUE, ncol = 3, driver_name = drivers_onechamp)
onechamp + ylim(-1, 6) + xlim(1980, 2019)
```

```r
drivers_multichamp <- c("Michael Schumacher","Ayrton Senna","Alain Prost","Lewis Hamilton")
multichamp <- plot_rating(rating,facet = TRUE, ncol = 2, driver_name = drivers_multichamp)
multichamp + ylim(-1, 6) + xlim(1980, 2019)
```
